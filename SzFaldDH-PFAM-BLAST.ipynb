```python
'''make BLAST database from FASTA file of InterPro protein family glutathione independent formaldehyde dehydorgenases'''
!makeblastdb -in protein-matching-IPR014184.fasta -dbtype prot -out IPR014184_db -parse_seqids
```

```python
!blastp -query seed.fasta -db IPR014184_db \
  -outfmt "6 qseqid sseqid pident length mismatch gapopen qstart qend sstart send evalue bitscore" \
  -evalue 1e-5 -max_target_seqs 500 -num_threads 4 -out SzFaldDH-PFAM-blast_results.tsv
```

```python
import pandas as pd
cols = ["qseqid","sseqid","pident","length","mismatch","gapopen","qstart","qend","sstart","send","evalue","bitscore"]
df = pd.read_csv("SzFaldDH-PFAM-blast_results.tsv", sep="\t", names=cols)
df = df.sort_values(["evalue","bitscore"], ascending=[True,False])
display(df.head(50))
```

```python
cut = df[(df["evalue"] <= 1e-5) & (df["pident"] >= 50)]
cut = cut.sort_values(["evalue","bitscore"], ascending=[True,False])
SzFaldDH_PFAM_best_ids = cut["sseqid"].unique().tolist()

print(f"{len(SzFaldDH_PFAM_best_ids)} sequences passed the cutoff")
```

```python
from Bio import SeqIO
records = SeqIO.to_dict(SeqIO.parse("protein-matching-IPR014184.fasta", "fasta"))

with open("SzFaldDH_PFAM_blast_hits_filtered.fasta", "w") as out:
    for hid in SzFaldDH_PFAM_best_ids:
        if hid in records:
            SeqIO.write(records[hid], out, "fasta")
```
